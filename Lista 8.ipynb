## üìä **Exerc√≠cio 1: An√°lise Dual com Gr√°ficos de Barras e Dispers√£o**

### O que voc√™ vai fazer:
1. Crie dois gr√°ficos em uma √∫nica visualiza√ß√£o:
  - Um **gr√°fico de barras** para mostrar as vendas mensais de uma empresa ao longo de um ano.
  - Um **gr√°fico de dispers√£o** para mostrar a rela√ß√£o entre as vendas mensais e os meses do ano.

   ```plaintext
   Meses: ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
   Vendas: [2100, 2600, 2800, 2300, 2700, 3100, 3300, 3200, 2900, 3000, 3400, 3500]
   ```

### Adicione:
- No gr√°fico de barras:
  - Um t√≠tulo: `"Vendas Mensais de 2024"`.
  - R√≥tulos no eixo **x** para os meses e no eixo **y** para o valor das vendas.

- No gr√°fico de dispers√£o:
  - R√≥tulos para os eixos: `"Meses"` e `"Vendas (R$)"`.
  - Um destaque em uma cor √∫nica para o m√™s com a maior venda.

### Dica:
Use cores diferentes para os dois gr√°ficos e ajuste as escalas para garantir clareza visual.


````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
````

````python
# --- Exerc√≠cio 1 ---

# Dados
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
vendas = [2100, 2600, 2800, 2300, 2700, 3100, 3300, 3200, 2900, 3000, 3400, 3500]

df = pd.DataFrame({'Meses': meses, 'Vendas': vendas})

##### SEU C√ìDIGO AQUI #####

#Configura√ß√£o da figura
plt.figure(figsize=(14, 6))

# Cria√ß√£o do Gr√°fico de Barras
plt.subplot(1, 2, 1)
plt.bar(df['Meses'], df['Vendas'], color='dodgerblue')
plt.title('Vendas Mensais de 2024')
plt.xlabel('Meses')
plt.ylabel('Valor das Vendas (R$)')

# Cria√ß√£o do Gr√°fico de dispers√£o
plt.subplot(1, 2, 2)
plt.scatter(df['Meses'], df['Vendas'], color='tomato', marker='s')
plt.title('Vendas Mensais de 2024')
plt.xlabel('Meses')
plt.ylabel('Vendas (R$)')
plt.scatter(meses[vendas.index(max(vendas))], max(vendas), color='darkgreen', zorder=5)

plt.tight_layout()
plt.show()
````

## üìà **Exerc√≠cio 2: Gr√°fico de Barras e Histograma**

### O que voc√™ vai fazer:
1. Crie dois gr√°ficos em uma √∫nica visualiza√ß√£o:
    - Um **gr√°fico de barras** que compare as despesas mensais por categoria.
    - Um **histograma** que mostre a distribui√ß√£o das idades dos funcion√°rios de uma empresa.

     ```plaintext
     Categorias: ['Alimenta√ß√£o', 'Transporte', 'Sa√∫de', 'Educa√ß√£o', 'Lazer']
     Valores (R$): [1200, 500, 300, 800, 200]
     Idades: [23, 25, 22, 30, 35, 40, 29, 28, 45, 50, 24, 27, 32, 33, 48, 55, 25, 26, 31, 37]
     ```

### Adicione:
- No gr√°fico de barras:
  - Um t√≠tulo: `"Despesas Mensais por Categoria"`.
  - R√≥tulos no eixo **x** para as categorias e no eixo **y** para os valores das despesas.

- No histograma:
  - Um t√≠tulo: `"Distribui√ß√£o das Idades dos Funcion√°rios"`.
  - R√≥tulos para os eixos: `"Idades"` e `"Frequ√™ncia"`.

````python
# dados
categorias = ['Alimenta√ß√£o', 'Transporte', 'Sa√∫de', 'Educa√ß√£o', 'Lazer']
valores = [1200, 500, 300, 800, 200]
idades = [23, 25, 22, 30, 35, 40, 29, 28, 45, 50, 24, 27, 32, 33, 48, 55, 25, 26, 31, 37]

df = pd.DataFrame({'Categorias': categorias, 'Valores': valores})
series_idades = pd.Series(idades)


##### SEU C√ìDIGO AQUI #####

#Configura√ß√£o
plt.figure(figsize=(12, 6))

#Gr√°fico de barrar (Despesas por Categoria)
plt.subplot(1, 2, 1)
plt.bar(df['Categorias'], df['Valores'], color='forestgreen')
plt.title('Despesas Mensais por Categoria')
plt.xlabel('Categorias')
plt.ylabel('Valores das Despesas (R$)')

#Gr√°fico de histograma (Distribui√ß√£o das Idades)
plt.subplot(1, 2, 2)
plt.hist(series_idades, bins=10, color='darkorange', edgecolor='midnightblue')
plt.title('Distribui√ß√£o das Idades dos Funcion√°rios')
plt.xlabel('Idades')
plt.ylabel('Frequ√™ncia')

plt.tight_layout()
plt.show()

````

## ü•ß **Exerc√≠cio 3: Gr√°fico de Pizza e Gr√°fico de Linha**

### O que voc√™ vai fazer:
1. Crie dois gr√°ficos em uma √∫nica visualiza√ß√£o:
  - Um **gr√°fico de pizza** que mostre a participa√ß√£o de mercado de diferentes empresas no setor de tecnologia.
  - Um **gr√°fico de linha** que mostre a temperatura ao longo da semana.

  ```plaintext
  Empresas: ['Apple', 'Samsung', 'Google', 'Microsoft']
  Participa√ß√£o (%): [40, 30, 20, 10]
  Dias: ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'S√°b', 'Dom']
  Temperatura (¬∞C): [22, 24, 19, 23, 25, 20, 21]
  ```

### Adicione:
- No gr√°fico de pizza:
  - Um t√≠tulo: `"Participa√ß√£o de Mercado no Setor de Tecnologia"`.
  - R√≥tulos para cada fatia com o nome da empresa e a porcentagem correspondente.

- No gr√°fico de linha:
  - Um t√≠tulo: `"Temperatura ao Longo da Semana"`.
  - R√≥tulos no eixo **x** para os dias e no eixo **y** para a temperatura em ¬∞C.
  - Destaque o dia com a temperatura mais alta em uma cor diferente.

````python
# Dados
empresas = ['Apple', 'Samsung', 'Google', 'Microsoft']
participacao = [40, 30, 20, 10]
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'S√°b', 'Dom']
temperatura = [22, 24, 19, 23, 25, 20, 21]

df_empresas = pd.DataFrame({'Empresas': empresas, 'Participa√ß√£o': participacao})

df_dias = pd.DataFrame({'Dias': dias, 'Temperatura': temperatura})

##### SEU C√ìDIGO AQUI #####

#Configura√ß√£o
plt.figure(figsize=(14, 6))

#Gr√°fico de Pizza
plt.subplot(1, 2, 1)
plt.pie(df_empresas['Participa√ß√£o'], labels=df_empresas['Empresas'], autopct='%1.1f%%', startangle=90, colors=['silver', 'springgreen', 'orange', 'skyblue'])
plt.title('Participa√ß√£o de Mercado no Setor de Tecnologia')

#Gr√°fico de Linha
plt.subplot(1, 2, 2)
plt.plot(df_dias['Dias'], df_dias['Temperatura'], marker='o', color='gold', linestyle='-', linewidth=2)
plt.title('Temperatura ao Longo da Semana')
plt.xlabel('Dias')
plt.ylabel('Temperatura (¬∞C)')
#Destaque para a temperatura mais alta
plt.scatter(dias[temperatura.index(max(temperatura))], max(temperatura), color='r', zorder=5)

plt.tight_layout()
plt.show()
````

## üìâ **Exerc√≠cio 4: Gr√°fico de Barras e Dispers√£o com Seaborn**

### O que voc√™ vai fazer:

1. Crie dois gr√°ficos em uma √∫nica visualiza√ß√£o:
    - Um **gr√°fico de barras** que mostre a m√©dia de vendas por categoria de produto.
    - Um **gr√°fico de dispers√£o** que mostre a rela√ß√£o entre altura e peso de um grupo de pessoas.

    ```plaintext
     Categorias: ['A', 'B', 'C', 'D']
     Vendas: [100, 150, 300, 250]
     Altura (cm): [150, 160, 170, 180, 190]
     Peso (kg): [50, 60, 70, 80, 90]
    ```

### Adicione:
- No gr√°fico de barras:
  - Um t√≠tulo: `"M√©dia de Vendas por Categoria"`.
  - R√≥tulos no eixo **x** para as categorias e no eixo **y** para os valores das vendas.

- No gr√°fico de dispers√£o:
  - R√≥tulos para os eixos: `"Altura (cm)"` e `"Peso (kg)"`.
  - Um destaque em uma cor √∫nica para o ponto mais alto e pesado.

````python
# Dados
categorias = ['A', 'B', 'C', 'D']
vendas = [100, 150, 300, 250]
altura = [150, 160, 170, 180, 190]
peso = [50, 60, 70, 80, 90]

df_vendas = pd.DataFrame({'Categorias': categorias, 'Vendas': vendas})

df_pessoas = pd.DataFrame({'Altura': altura, 'Peso': peso})

##### SEU C√ìDIGO AQUI #####

#Configura√ß√£o
plt.figure(figsize=(14, 6))

#Gr√°fico de Barras
plt.subplot(1, 2, 1)
sns.barplot(x='Categorias', y='Vendas', data= df_vendas, hue='Categorias', palette='deep', dodge=False, legend=True)
plt.title('M√©dia de Vendas por Categoria')
plt.xlabel('Categorias')
plt.ylabel('Valores de Vendas')

#Gr√°fico de Dispers√£o
plt.subplot(1, 2, 2)
sns.scatterplot(x='Altura', y='Peso', data= df_pessoas, s= 100, color= 'blue')
plt.title('Altura vs Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.scatter(df_pessoas.iloc[-1, 0], df_pessoas.iloc[-1, 1], color='red', zorder=5)

plt.tight_layout()
plt.show()

````

## üî• **Exerc√≠cio 5: Histograma e Heatmap com Seaborn**

### O que voc√™ vai fazer:

1. Crie dois gr√°ficos em uma √∫nica visualiza√ß√£o:
    - Um **histograma** que mostre a distribui√ß√£o das idades dos participantes de um estudo.
    - Um **heatmap** que mostre a correla√ß√£o entre as vendas de diferentes produtos.

    ```plaintext
    Idades: [22, 25, 27, 29, 32, 35, 38, 40, 42, 45, 48, 50]
    Produto A: [100, 150, 120, 130, 140, 160, 180]
    Produto B: [80, 90, 110, 120, 115, 130, 140]
    Produto C: [60, 75, 90, 100, 105, 110, 120]
    ```

### Adicione:
- No histograma:
  - Um t√≠tulo: `"Distribui√ß√£o das Idades dos Participantes"`.
  - R√≥tulos para os eixos: `"Idades"` e `"Frequ√™ncia"`.

- No heatmap:
  - Um t√≠tulo: `"Correla√ß√£o entre Vendas de Diferentes Produtos"`.
  - R√≥tulos para os eixos: `"Produtos"` e `"Vendas"`.

````python
import pandas as pd

# Dados
idades_estudo = [22, 25, 27, 29, 32, 35, 38, 40, 42, 45, 48, 50]
produto_a = [100, 150, 120, 130, 140, 160, 180]
produto_b = [80, 90, 110, 120, 115, 130, 140]
produto_c = [60, 75, 90, 100, 105, 110, 120]

series_idades = pd.Series(idades_estudo)

vendas_produtos = pd.DataFrame({
    'Produto A': produto_a,
    'Produto B': produto_b,
    'Produto C': produto_c
})

##### SEU C√ìDIGO AQUI #####

#Configura√ß√£o
plt.figure(figsize=(14, 6))

#Gr√°fico de Histograma
plt.subplot(1, 2, 1)
sns.histplot(series_idades, bins=10, color='lightblue', kde=True)
plt.title('Distribui√ß√£o das Idades dos Participantes')
plt.xlabel('Idades')
plt.ylabel('Frenqu√™ncia')

#Gr√°fico de Heatmap
plt.subplot(1, 2, 2)
sns.heatmap(vendas_produtos.corr(), annot= True, cmap='coolwarm', cbar= True)
plt.title('Correla√ß√£o entre Vendas de Diferentes Produtos')
plt.xlabel('Produtos')
plt.ylabel('Vendas')

plt.tight_layout()
plt.show()
````
